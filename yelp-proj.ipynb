```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sb
```

```python
bdf = pd.read_csv("yelp-proj-data/business_cl.csv")
bdf = bdf.drop("business_id", axis=1)
```

```python
corr = bdf.corr()
fig, ax = plt.subplots(figsize=(20,20))
sb.heatmap(corr, linewidths=.3, cmap="YlGnBu", ax=ax)
```
